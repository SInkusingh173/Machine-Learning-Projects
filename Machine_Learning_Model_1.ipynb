# ** My first Machine learning Project **


## Load Data


```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

##Data Preparation

### Data Preparation  X and Y

```python
y = df['logS']
y
```

```python
x = df.drop('logS' , axis = 1)
x
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train, y_test = train_test_split(x , y , test_size= 0.2 ,random_state= 100)
```

```python
x_train
```

```python
x_test
```

## ** Model Building **

### Linear Regression

##### Traning the **model

```python
## there is a finaly MAchine learing model created know for follwing LinearRegression  Algorithms
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

#### Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
print(y_lr_train_pred, y_lr_test_pred)
```

```python
## 80% of training data
y_lr_train_pred
```

```python
##20% of  testing data
y_lr_test_pred
```

## Evaluate Model Performing

```python
## data actual value
y_train
```

```python
##Giving data Prediting values
y_lr_train_pred
```

###Actual Model Evalurating Sectin

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

##Printing actual Model values

```python
print('lr mse (Train) -->', lr_train_mse)
print('lr r2 (Train) -->', lr_train_r2)

print('lr mse (Test) -->', lr_test_mse)
print('lr r2 (Test) -->', lr_test_r2)
```

```python
##Showing result
lr_results = pd.DataFrame(['Linear regression' , lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
```

```python
##final data pred values
lr_results
```

```python
lr_results.columns = ['Method' , ' Training MSE ' , ' Training R2' , ' Test MSE' , ' Test R2' ]
```

## Data Visualization of prefication results

```python
import matplotlib.pyplot as plt
import numpy as np
##using alpha for overlapping Points
## c is graph colour code
## y label is a graph   Predict axis name
## x label is a graph  experimental axis name


plt.scatter(x=y_train, y=y_lr_train_pred  , alpha= 0.3 , c="#7CAE00" )


## z axis line drowing
z = np.polyfit(y_train, y_lr_train_pred , 1)
p = np.poly1d(z)

##red line in z axis

plt.plot(y_train, p(y_train) , '#F8766D')

plt.plot()
plt.ylabel('Predict Logs')
plt.xlabel('Experimental logs')
plt.figure(figsize= (5,5))
```
